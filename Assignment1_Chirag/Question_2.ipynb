Question.2 This question is for familiarity with strings. You are needed to input a string and check if it is a palindrome.

```python
# We can define a function which checks if a string is in a palindrome
def isPalindrome(str):
  left = 0
  right = len(str)-1
  while left < right:
    if str[left] == str[right]:
      left += 1
      right -= 1
      return True
    else:
      return False
input_str = input("Enter a string: ")
if isPalindrome(input_str):
  print("The string is a Palindrome")
else:
  print("The string is not a Palindrome")

```
