1
a1, a2, . . . , an, and a number x. You need to find a pair of elements whose absolute difference
is x. You are assured that such a pair exists in the array

```python
# Given an array of n integers and a number(target)
# We can use the two pointer method for that first we need to sort the array
# We can define function to sort an array and then finding the pair

def find_pair(arr, target):
  arr.sort()
  n = len(arr)
  req_pair = []
  i, j = 0, 1
  while j < n:
    diff = arr[j] - arr[i]
    if diff == target:
      current_i = arr[i]      # To set the values of i and j at which difference requirement is met
      current_j = arr[j]
      k = i
      while k < n and arr[k] == current_i:       # Inner loop for finding duplicates
        l = j
        while l < n and arr[l] == current_j:
          req_pair.append((arr[i],arr[j]))
          l +=1
        k +=1
      while i < n and arr[i] == current_i:      # To skip all the duplicates i values
        i += 1
      while j < n and arr[j] == current_j:
        j += 1
    elif diff < target:
      j += 1
    else:
      i += 1
  return req_pair
unsorted_arr = [3, 5, 6, 3, 9, 7]
target = 2
find_pair(unsorted_arr, target)








```
