

Question 1. The first question involves arithmetic with a small application of loops. You are given an
array of n integers, that is, a1, a2, . . . , an, and you are required to output the maximum and
minimum among them.



```
# Given an array or list of n numbers
arr = [ 3, 5, 56, -2, 78, 100]
max=arr[0]
min=arr[0]
i=0
while i < len(arr):
  if arr[i]>max:        # For maximum
    max = arr[i]
  if arr[i]<min:        # For minimum
    min = arr[i]
  i+=1
print(f"The maximum value is {max}")    # Simiilar can done for an array of n integers
print(f"The minimum value is {min}")
```
